https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing first data cleaning, formatting, and structuring.

Start by reading the dataset, standardizing column names, dealing with null values and creating the month column. You can do so by using the functions you developed in previous labs, or by reading the clean dataset you saved in the last lab `data structuring and combining`.

```python
import pandas as pd
import numpy as np
import datetime as dt

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df = pd.read_csv(url)
```

```python
df.head(2)
```

```python
# Standardize column names
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Deal with null values
df = df.dropna()

# Create the month column
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])
df['month'] = df['effective_to_date'].dt.month

# Display the first few rows of the dataframe
df.head()
```

Then, answer the following questions:

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
# Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000
# and have a response of 'Yes' to the last marketing campaign
df_filtered = df[(df['total_claim_amount'] > 1000) & (df['response'] == 'Yes')]

# Display the first few rows of the new dataframe
df_filtered.head()
```

2. Analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign.

```python
# Analyze the average total_claim_amount by each policy type and gender
# for customers who have responded 'Yes' to the last marketing campaign
average_claim = df[df['response'] == 'Yes'].groupby(['policy_type', 'gender'])['total_claim_amount'].mean()
average_claim
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Analyze the total number of customers who have policies in each state
state_counts = df['state'].value_counts()

# Filter the results to only include states where there are more than 500 customers
state_counts = state_counts[state_counts > 500]
state_counts
```

4. Find the median customer lifetime value by education level and gender.

```python
# Find the median customer lifetime value by education level and gender
median_value = df.groupby(['education', 'gender'])['customer_lifetime_value'].median()
median_value
```

5. Find the maximum, minimum, and average number of open complaints by state and policy type.

```python
# Find the maximum, minimum, and average number of open complaints by state and policy type
complaints_stats = df.groupby(['state', 'policy_type'])['number_of_open_complaints'].agg(['max', 'min', 'mean'])
complaints_stats
```

6. The marketing team wants to analyze the number of policies sold by state and month.

Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
# The marketing team wants to analyze the number of policies sold by state and month
# Group the data by state and month, then count the number of policies sold for each group
policies_sold = df.groupby(['state', 'month'])['policy_type'].count()

# Present the data in a table where the months are arranged as columns and the states are arranged as rows
policies_sold = policies_sold.unstack(level=-1)
policies_sold
```

7. Display a new DataFrame that contains the number of policies sold by month by state for the top 3 states with the highest number of policies sold.

*Hint:* 
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
# Display a new DataFrame that contains the number of policies sold by month by state
# for the top 3 states with the highest number of policies sold

# Calculate total policies sold by state
total_policies_sold = df['state'].value_counts()

# Get the top 3 states with the highest number of policies sold
top_states = total_policies_sold.nlargest(3).index

# Filter the policies_sold DataFrame to only include the top 3 states
top_states_policies_sold = policies_sold[policies_sold.index.isin(top_states)]
top_states_policies_sold
```

8. The marketing team wants to analyze the effect of different marketing channels on the customer response rate.

Hint: You can use melt to unpivot the data and create a table that shows the customer response rate (those who responded "Yes") by marketing channel.

```python
# The marketing team wants to analyze the effect of different marketing channels on the customer response rate
# Create a table that shows the customer response rate (those who responded 'Yes') by marketing channel

# Calculate the total number of responses by sales channel
total_responses = df['sales_channel'].value_counts()
total_responses
```

```python
# Calculate the number of 'Yes' responses by sales channel
yes_responses = df[df['response'] == 'Yes']['sales_channel'].value_counts()
yes_responses
```

```python
# Calculate the response rate by sales channel
response_rate = yes_responses / total_responses
response_rate
```

```python
# Convert the response rate series to a DataFrame
response_rate_df = response_rate.reset_index()
response_rate_df.columns = ['sales_channel', 'response_rate']
response_rate_df
```
